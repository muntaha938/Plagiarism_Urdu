# Dataset

**RESTART RUNTIME**

```
import pandas as pd 
from ast import literal_eval
df=  pd.read_csv("Final_data.csv", encoding='utf-8', header=0, converters={"Sentence": literal_eval, "Sentiment": literal_eval})
print(df.head())
df['Sentence'] = df['Sentence'].map(lambda l: l[0])
df['Sentiment'] = df['Sentiment'].map(lambda l: l[0])
print(df.head())
```

```
df['Sentiment'].value_counts()
```

```
from sklearn import preprocessing
labeler = preprocessing.LabelEncoder()

from sklearn.model_selection import train_test_split
df = df.rename(columns={'Sentence': 'text', 'Sentiment': 'labels'})
df['labels'] = labeler.fit_transform(df['labels'])
df.head()
```

```
df_train, df_test = train_test_split(df, test_size=0.2)
```

# mBERT standard

```
# set use_cuda=False on CPU-only platforms
model = ClassificationModel('bert', 'bert-base-multilingual-cased', num_labels=3, use_cuda=True, args={
    'reprocess_input_data': True,
    'num_train_epochs': 30,
})
model.train_model(df_train, eval_df=df_test)
```

```
result, model_outputs, wrong_predictions = model.eval_model(df_test)
bads = {}
for pred in wrong_predictions:
    if pred.label in bads:
        bads[pred.label] += 1
    else:
        bads[pred.label] = 1
print("Incorrect labels")
print(bads)
```

```
print(result)
print(model_outputs)
print(df_test)
```
